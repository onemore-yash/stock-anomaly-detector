```python
#Calculate daily price change(percentage)
data['Price_Change_%']=data['Close'].pct_change()*100

print("="*50)
print("CAILY PRICE CHANGES")
print("="*50)
print(data[['Close','Price_Change_%']].head(10))

print("\n"+"="*50)
print("STATISTICS OF PRICE CHANGES")
print("="*50)
print(data['Price_Change_%'].describe())


print("\n"+"="*50)
print("TOP 10 BIGGEST PRICE INCREASES")
print("="*50)
print(data['Price_Change_%'].nlargest(10))

print("\n"+"="*50)
print("TOP 10 BIGGEST PRICE DROPS")
print("="*50)
print(data['Price_Change_%'].nsmallest(10))

#Visualize price changes over time
plt.figure(figsize=(14,6))
plt.plot(data.index,data['Price_Change_%'], linewidth=1,color='green')
plt.axhline(y=0,color='black',linestyle='-',linewidth=0.5)
plt.axhline(y=data['Price_Change_%'].mean(),color='red',linestyle='--',label='Mean')
plt.axhline(y=data['Price_Change_%'].mean()+data['Price_Change_%'].std(),color='orange',linestyle='--',label='Mean + 1 Std')
plt.axhline(y=data['Price_Change_%'].mean()-data['Price_Change_%'].std(),color='orange',linestyle='--',label='Mean - 1 Std')
plt.title('Daily Price Change (%) - Apple Stock',fontsize=14,fontweight='bold')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Price Change %',fontsize=12)
plt.legend()
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```
